```python
import pandas as pd
import math
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv/carprices.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model']).astype(int)
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
final = merged.drop(columns=['Car Model', 'Mercedez Benz C class'])
final
```

```python
model = LinearRegression()
X = final.drop(columns='Sell Price($)')
y = final['Sell Price($)']
model.fit(X, y)

predictions = model.predict([[86000, 7, 0, 1], [50000, 5, 1, 0], [45000, 4, 0, 0]])
print(f'BMW X5 (mileage=50k, 5yrs old): {math.floor(predictions[0])}$')
print(f'Audi A5 (mileage=86k, 7yrs old): {math.floor(predictions[1])}$')
print(f'Mercedez Benz C class (mileage=45k, 4yrs old): {math.floor(predictions[2])}$')
print(f'Model accuracy = {model.score(X, y)}')
```
